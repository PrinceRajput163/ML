```python
#Prince Rajput
#2400290120190
#CS4C
#LAb = 4
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
df = pd.read_csv('Iris.csv')
df = df.drop(columns=['Id'])
df
```

```python
x = df.drop(columns=['Species'])
y = df['Species']
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,
test_size = 0.3,
random_state=42,
stratify=y)
df
```

```python
knn = KNeighborsClassifier(
    n_neighbors=5,
    metric="euclidean",
    weights="uniform"
)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
print("KNN Aaccuracy:",accuracy_score(y_test, y_pred_knn))
print("KNN Confusion Matrix:\n",confusion_matrix(y_test, y_pred_knn))
```

```python
wknn = KNeighborsClassifier(
    n_neighbors=10,
    metric="euclidean",
    weights="distance"
)

wknn.fit(x_train, y_train)

y_pred_wknn = wknn.predict(x_test)

print("WKNN Accuracy:", accuracy_score(y_test, y_pred_wknn))
print("WKNN Confusion Matrix:\n", confusion_matrix(y_test, y_pred_wknn))
```
